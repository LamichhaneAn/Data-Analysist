```python
#import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Create the dataframe:

import pandas as pd

# Read the CSV file into a DataFrame
df = pd.read_csv("IMDB-Movie-Data.csv")

# Display the first 5 rows of the DataFrame
df.head(5)
```

```python
df.shape
```

```python
#To get information about dataset
df.info()
```

```python
#Check the missing values in the dataset

df.isnull().sum()
```

```python
#alternative method to find missing value

import seaborn as sns
import matplotlib.pyplot as plt

# Create a heatmap to visualize missing values
sns.heatmap(df.isnull())

# Display the plot
plt.show()
```

```python
#What is the percentage of getting missing values:

percentage_missing=df.isnull().sum()*100/len(df)
percentage_missing
```

```python
#drop all the missing values: How to drop?using drop function:

df.dropna(axis = 0)

#axis = 0 meant to drop the rows
#axis = 1 meant to drop the column
#N plus = meant to modify the state of a net offer instead of an existing dataframe.
```

```python
#check for duplicate data

duplicate_data=df.duplicated().any()
duplicate_data
```

If there were any duplicate valus than we have to use this syntax:

df=df.drop_duplicates()

```python
#To get statistics about the dataframe:

df.describe(include="all")
```

```python
df.columns
```

```python
#display the title of the movie having runtime >= 180minutes

df[df["Runtime (Minutes)"]>=180]["Title"]
```

```python
#I want to display In which year there was the highest average voting:

df.groupby("Year")["Votes"].mean()
```

```python
df.groupby("Year")["Votes"].mean().sort_values(ascending=False)
```

```python
#lets create the plot for highest average voting:

sns.barplot(x="Year", y="Votes", data=df)
plt.title("Average votes by year")
plt.show()
```

```python
#In which year there was the highest average revenue:

df.groupby("Year")["Revenue (Millions)"].mean()
```

```python
df.groupby("Year")["Revenue (Millions)"].mean().sort_values(ascending=False)
```

```python
#create a plot

sns.barplot(x = "Year", y = "Revenue (Millions)", data = df)
plt.title("Highest Average Revenue by the Year")
plt.show()
```

```python
#I want to find the average rating of each director

df.groupby("Director")["Rating"].mean()
```

```python
df.groupby("Director")["Rating"].mean().sort_values(ascending=False)
```

```python
# Sort the DataFrame by 'Runtime' in descending order
top_10_lengthy_movies = df.sort_values(by='Runtime (Minutes)', ascending=False).head(10)

# Select and display only the 'Title' and 'Runtime (Minutes)' columns
top_10_lengthy_movies[['Title', 'Runtime (Minutes)']]
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Creating a bar plot
sns.barplot(x="Runtime (Minutes)", y="Title", data=top_10_lengthy_movies, palette="viridis")

# Display the plot
plt.show()
```

```python
#I want to create number of movies per year

df["Year"].value_counts()
```

```python
sns.countplot(x = "Year", data=df)
plt.show()
```

```python
#I wan to find the most popular movie title (based on the highest revenue)

df["Revenue (Millions)"].max()
```

```python
df[df["Revenue (Millions)"].max()==df["Revenue (Millions)"]]["Title"]
```

```python
# Example: Calculate the average rating of movies per genre
avg_rating_per_genre = df.groupby('Genre')['Rating'].mean()
avg_rating_per_genre
```

```python
# Example: Count the number of movies per year
movies_per_year = df.groupby('Year').size()
movies_per_year
```

```python
# Example: Filter out directors who have directed more than 5 movies
directors_with_more_than_5_movies = df.groupby('Director').filter(lambda x: len(x) > 5)
directors_with_more_than_5_movies
```
